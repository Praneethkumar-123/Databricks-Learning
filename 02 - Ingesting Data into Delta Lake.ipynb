#COPY INTO

- create a table from employee,csv file using COPY INTO Statement
- THE COPY INTO statement incrementally loads data from a file location into delta table. 
- This is recyclable and idempotent operation. 
- Files in source location that have already been loaded are skipped.

```python
%sql
drop table if exists current_customer_copyinto;
```

```python
%sql
create table current_customer_copyinto(
  customer_id varchar(10),
  customer_name varchar(50),
  email varchar(50),
  phone varchar(20),
  address varchar(100),
  country varchar(20),
  is_prime_member boolean
);
```

```python
%sql
copy into current_customer_copyinto
FROM '/Volumes/workspace/labuser1/myfiles/'
FILEFORMAT=CSV
FORMAT_OPTIONS (
  'header'='true',
  'inferschema'='true'
)
```

```python
%sql
select * from current_customer_copyinto;
```

```python
%sql
copy into current_customer_copyinto
FROM '/Volumes/workspace/labuser1/myfiles/'
FILEFORMAT=CSV
FORMAT_OPTIONS (
  'header'='true',
  'inferschema'='true'
)
```
