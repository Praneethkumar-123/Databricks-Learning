```python
%sql
drop table if exists customers;
create table customers as
select * from read_files(
  '/Volumes/workspace/'||'/labuser1/'||'/myfiles/',
  format=>'csv',
  header=>true,
  inferschema=>true
);

select * from customers;
```

```python
#Read the csv and create a spark dataframe
df=(spark
    .read
    .format("csv")
    .option("header",'true')
    .option("inferschema",'true')
    .load(f'/Volumes/workspace/labuser1/myfiles/')
    )

#create a delta table from spark data frame
(df
.write
.mode("overwrite")
.format("delta")
.saveAsTable(f"workspace.labuser1.customer"))
```
